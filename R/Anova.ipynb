```r
tratamento = read.csv(file.choose(), sep = ";", header=T) # File anova.csv
fix(tratamento)
```

```r
boxplot(tratamento$Horas ~ tratamento$Remedio)
```

```r
an = aov(Horas ~ Remedio, data=tratamento)
summary(an)
```

```r
an = aov(Horas ~ Remedio * Sexo, data=tratamento)
summary(an)
```

```r
an = aov(Horas ~ Remedio, data=tratamento)
tukey = TukeyHSD(an)
tukey
```

```r
plot(tukey)
```
