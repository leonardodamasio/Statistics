```python
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
dados = norm.rvs(size=100) # rvs = Random variates | Gerar dados em uma distribuição normal
dados
```

```python
stats.probplot(dados, plot=plt)
```

```python
stats.shapiro(dados) # Teste de Shapiro-Wilk
```
